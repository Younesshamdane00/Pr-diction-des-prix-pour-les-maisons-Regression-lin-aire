### PreProcessing

```python
#Pour supprimer les valeurs extreme, on applique le log(salePrice)
df['SalePrice'].plot.hist(bins=100)
```

```python
#Pour supprimer les valeurs extreme, on applique le logarithme(salePrice)
df['SalePricelog'] = np.log(df.SalePrice)
df['SalePricelog'].plot.hist(bins=100)
```

```python
#Traitement pour le champs : Heating ( Type de chauffage ) : valeur catÃ©gorical
df.Heating.value_counts()
```

```python
#On va transformer les valeurs catÃ©gorique vers des valeurs numÃ©rique
cat_variables = df.select_dtypes(include=['object']).columns
cat_df        = pd.get_dummies(df[cat_variables])
cat_df.head()
```

```python
#Valeurs numÃ©riques
num_df = df[[c for c in df if c not in cat_variables]]
num_df
```

```python
#Concatenation des valeurs categorique + numÃ©rique
joined_df = pd.concat([cat_df,num_df],axis=1)
joined_df.head(3).T
```

```python
#Chaque champs : combien de valeurs null
df.isna().sum(axis=0).sort_values(ascending=False)
```

```python
#Remplissage des valeurs null avec la median de la variable
num_df.median()
```

### Simple regression lineaire

```python
joined_df.sample(3)
```

```python
#On commence avec ces deux valeurs car ils avaient une grande correlation avec le prix
features_df = joined_df[['OverallQual','GrLivArea']]
target      = joined_df.SalePricelog         
features_df.sample(3)
```

```python
#Training
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(features_df, target)
```

```python
#Prediction
predictions = model.predict(features_df)
predictions
```

```python
#MSE  = PREDICTION - TARGET =  np.sqrt(((predictions - target) **2).sum())
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(predictions,target))
```

```python
#Comparer le model avec un classifieur bete
from sklearn.dummy import DummyRegressor
baseline    = DummyRegressor('mean')
baseline.fit(features_df,target)
np.sqrt(mean_squared_error(baseline.predict(features_df),target))
```

```python
plt.scatter(target,predictions)
plt.plot([10., 14.],[10.,14.],'--') #Fonction d'un predicteur parfait
```

### Modeling avec la division du dataset avec la normalisation des donnÃ©es avec deux variables

```python
#On va diviser le dataset ( training / test = 20%)
# X entree de model // y sortie de model
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(features_df, target, test_size=0.25,random_state=0)
```

```python
#Training
from sklearn.linear_model import LinearRegression, LassoCV
# import LinearRegression : nomral regression lineaire
# import LassoCV : Regression lineaire regularisÃ©
model = LinearRegression(normalize=False)
model.fit(x_train, y_train)
```

```python
#Prediction
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,y_pred)) 
```

```python
#Comparer le model avec un classifieur simple
from sklearn.dummy import DummyRegressor
baseline    = DummyRegressor('mean')
baseline.fit(x_train, y_train)
np.sqrt(mean_squared_error(y_test,baseline.predict(x_test)))
```

```python
plt.scatter(y_test, y_pred,c='g') #test set en vert
plt.scatter(y_train, model.predict(x_train),c='r',alpha=0.2) #train set en vert

plt.xlabel('Sale Price')
plt.ylabel('Sale Price')
plt.plot([10., 14.],[10.,14.],'--') #Fonction d'un predicteur parfait
```

### Modeling avec la division du dataset avec plusieurs variables

```python
features_df = joined_df.copy().fillna(0.0)
target      = features_df.pop('SalePrice')     
features_df.sample(3)
```

```python
#( training / test = 75%, 25%)
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(features_df, target, test_size=0.25,random_state=0)
```

```python

from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV
# import LinearRegression : nomral regression lineaire
# import LassoCV : Regression lineaire regularisÃ©
model = LinearRegression(normalize=False)
model.fit(x_train, y_train)
```

```python
#Prediction
y_pred = model.predict(x_test)
```

```python
#Min squared error
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
#Comparaison entre test et predictions
resultat = pd.DataFrame()
resultat['test'] = y_test
resultat['pred'] = y_pred
resultat.sample(6)
```
